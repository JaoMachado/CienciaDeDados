Fazer um código que calcule a variância e o desvio padrão (sem utilizar funções prontas) dos salários da empresa apresentados no slide 6.
Calcule também a moda e mediana dos salários.
Apresente ao final do código a média e o desvio padrão dos salários da companhia.
Use um vetor para armazenar a lista de salários da empresa


```python
salarios = [45000, 15000, 10000, 10000, 5700, 5000, 5000, 5000, 3700, 3700, 3700, 3700, 3000]
ult = [2000 for a in range(12)]
```

```python
salarios.extend(ult)
```

```python
salarios
```

```python
media = np.mean(salarios)
```

Variância

```python
variancia = 0
for a in salarios:
  variancia += ((a - media)**2)

variancia = variancia/(len(salarios) - 1)
print("Variância: ", variancia)
```

Desvio Padrão

```python
desvioPadrao = variancia**(1/2)

print("Desvio Padrão: ", desvioPadrao)
```

Moda

```python
salModa = {}

for a in salarios:
  if a in salModa.keys():
    salModa[a] += 1
  else:
    salModa[a] = 1
salModa
```

```python
moda = max(salModa, key=salModa.get)
moda
```

Mediana

```python
Mediana = salarios[len(salarios)//2]
Mediana
```
