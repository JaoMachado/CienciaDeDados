```python
import pandas as pd
import matplotlib.pyplot as plt
```

Considere a Tabela A sobre o custo com propaganda e o lucro obtido por uma empresa em campanhas publicitárias.
Pesquise em Python como criar um DataFrame com os valores contidos em ambas colunas.
Use a função .plot() do DataFrame para gerar um gráfico com os valores da tabela.
Use a função .corr() do DataFrame para calcular a correlação de Pearson e Spearman entre ambas variáveis.
Há correlação significativa entre o custo com propaganda e o lucro obtido pela empresa?-> Sim, há uma forte correlação entre o custo com propaganda e o lucro obtivo.

```python
custo = [1000, 1150, 1500, 2000, 2200, 10000]
lucro = [5500, 5400, 5600, 5900, 6000, 20000]
```

```python
tabelaA = pd.DataFrame({'Custo com propaganda': custo, 'Lucro sobre cada item': lucro})
tabelaA
```

```python
tabelaA.plot()
```

```python
tabelaA.plot(kind='bar', x='Custo com propaganda', y='Lucro sobre cada item', title='Tabela A', figsize=(10, 5), color='blue')
```

```python
tabelaA.corr(method='pearson')
```

```python
tabelaA.corr(method='spearman')
```
