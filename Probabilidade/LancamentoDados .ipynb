Pesquise por uma biblioteca geradora de números aleatórios em Python e simule o lançamento de um dado de 6 faces por 100 vezes e depois por 1000 vezes. Verifique:
Quantas vezes caem cada face? R-> Respondido no código.
Neste dado, qual a probabilidade de ocorrer cada face, dado os resultados destes lançamentos? -> Respondido no código.
Com base no número médio que cada face ocorreu, pode-se dizer que este dado hipotético se aproxima de um dado honesto ou estaria viciado? R-> Pra 100 vezes, não. Para 10000, se aproxima, porém ainda não é totalmente honesto.


Jogando o dado 100 vezes

```python
results100 = []
```

```python
for a in range(100):
    num = random.randint(1,6)
    results100.append(num)
```

```python
np.array(results100)
```

```python
a, b = np.unique(results100, return_counts=True)
```

```python
for x in a:
  print(f'{x}: {b[x-1]}')
```

```python
for x in b:
  print(x / len(results100))
```

```python
print(1/6)
```
