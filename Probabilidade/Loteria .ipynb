Simulação de um Jogo de Loteria em Python
Um jogo de loteria consiste em escolher 6 números de um conjunto de 60. Você quer simular o sorteio de um bilhete e calcular a probabilidade de acertar 4, 5 ou todos os 6 números.
Tarefas:
Simular um sorteio de 6 números.
Calcular a probabilidade empírica de acertar exatamente 4, 5 ou 6 números ao longo de 1.000.000 de sorteios.

```python
numerosSorteados = random.sample(range(1, 61), 6)
numerosSorteados
```

```python
qtd_quatro = 0
qtd_cinco = 0
qtd_seis = 0

for x in range(1000000):
    numerosEscolhidos = random.sample(range(1, 61), 6)

    numerosComuns = set(numerosSorteados) & set(numerosEscolhidos)

    qtd_comuns = len(numerosComuns)

    if qtd_comuns == 4:
        qtd_quatro += 1
    elif qtd_comuns == 5:
        qtd_cinco += 1
    elif qtd_comuns == 6:
        qtd_seis += 1

print(f'Quantidade de jogos com 4 números: {qtd_quatro}')
print(f'Quantidade de jogos com 5 números: {qtd_cinco}')
print(f'Quantidade de jogos com 6 números: {qtd_seis}')
```

```python
prob_4 = qtd_quatro / 1000000
prob_5 = qtd_cinco / 1000000
prob_6 = qtd_seis / 1000000

print(f"Probabilidade de acertar 4 números: {prob_4:.6f}")
print(f"Probabilidade de acertar 5 números: {prob_5:.6f}")
print(f"Probabilidade de acertar 6 números: {prob_6:.8f}")
```
